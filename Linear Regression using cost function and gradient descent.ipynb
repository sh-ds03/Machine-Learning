```python
import numpy as np
import matplotlib.pyplot as plt

# Independent variable (input)
X = np.array([1, 2, 3, 4, 5])

# Dependent variable (output)
Y = np.array([2, 3, 5, 4, 6])
```

```python
# Calculate the mean of X and Y
mean_x_value = np.mean(X)
mean_y_value = np.mean(Y)
```

```python
# Number of observations
n = len(X)

# Calculating the numerator and denominator for the slope
numerator = 0
denominator = 0
```

```python
for i in range(n):
    numerator += (X[i] - mean_x_value) * (Y[i] - mean_y_value)
    denominator += (X[i] - mean_x_value) ** 2
```

```python
# Slope (m)
m = numerator / denominator
```

```python
# Intercept (b)
b = mean_y_value - (m * mean_x_value)
```

```python
# Regression line
regression_line = [(m * x) + b for x in X]
```

```python
# Plotting the data points and the regression line
plt.scatter(X, Y, color='blue', label='Data Points')
plt.plot(X, regression_line, color='red', label='Regression Line')
plt.xlabel('Independent Variable (X)')
plt.ylabel('Dependent Variable (Y)')
plt.legend()
plt.show()
```

```python
# Output the slope and intercept
print(f'Slope (m): {m}')
print(f'Intercept (b): {b}')
```

```python
# by using cost function and gradient descent 
import numpy as np
import matplotlib.pyplot as plt

# Independent variable (input)
X = np.array([1, 2, 3, 4, 5])

# Dependent variable (output)
Y = np.array([2, 3, 5, 4, 6])
```

```python
# cost function
def compute_cost(X, Y, m, b):
    n = len(Y)
    cost = (1/2*n) * np.sum((X*m + b - Y)**2)
    return cost
```

```python
# gradient descent
def gradient_descent(X, Y, m, b, learning_rate, iterations):
    n = len(Y)
    for _ in range(iterations):
        # Compute the partial derivatives
        m_gradient = (-2/n) * np.sum(X * (Y - (m*X + b)))
        b_gradient = (-2/n) * np.sum(Y - (m*X + b))
        
        # Update parameters
        m = m - learning_rate * m_gradient
        b = b - learning_rate * b_gradient
    
    return m, b
```

```python
m_initial = 0
b_initial = 0
learning_rate = 0.01
iterations = 1000
```

```python
m_optimal, b_optimal = gradient_descent(X, Y, m_initial, b_initial, learning_rate, iterations)
```

```python
regression_line = [(m_optimal * x) + b_optimal for x in X]

plt.scatter(X, Y, color='blue', label='Data Points')
plt.plot(X, regression_line, color='red', label='Regression Line')
plt.xlabel('Independent Variable (X)')
plt.ylabel('Dependent Variable (Y)')
plt.legend()
plt.show()

print(f'Optimal Slope (m): {m_optimal}')
print(f'Optimal Intercept (b): {b_optimal}')
```
